```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
from sklearn import preprocessing
from sklearn.decomposition import PCA
%matplotlib inline
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import cm
from scipy.interpolate import griddata
from matplotlib.ticker import LinearLocator, FormatStrFormatter
import sys
tf.reset_default_graph()
```

```python
true_theta = 2.
true_theta_0 = .5

x=np.linspace(-1,1,1000)
y=x**3+x**2*2+np.sin(10*x)+np.random.randn(x.size)*0.6

x = x.reshape((-1,1))
y = y.reshape((-1,1))
```

```python
total_generate = 1000
x1=np.linspace(2,5,total_generate).reshape((-1,1))
x2=np.linspace(0,10,total_generate).reshape((-1,1))
x3=np.random.rand(total_generate,1)
x4=np.random.rand(total_generate,1)
x = np.hstack((x1,x2,x3,x4))
print(x.shape)

y = np.linspace(0,1,total_generate).reshape((-1,1))
y = np.sin(10*y)*50
print(y.shape)

def normalization(x):
    min_max_scaler = preprocessing.MinMaxScaler()
    x_scaled_01 = min_max_scaler.fit_transform(x)
    x_scaled_std = preprocessing.scale(x)
    return x_scaled_01, x_scaled_std

x_scaled_01, x_scaled_std = normalization(x)
```

```python
def pca(data,component_ = 2):
    pca = PCA(n_components=component_)
    pca.fit(data)
    trans = pca.transform(data)
    print("total variance explained: ")
    total = 0
    for each in pca.explained_variance_ratio_:
        total += each
    print(total)
    return trans

def three_d_viz(trans,tat,way):
    x = trans[:,0].squeeze()
    y = trans[:,1].squeeze()
    z = tat.squeeze()
    
    #z[ z > 1 ] = 1
    #z[ z < 0 ] = 0

    
    xi = np.linspace(min(x), max(x))
    yi = np.linspace(min(y), max(y))
    Xg, Yg = np.meshgrid(xi, yi)
    Zg = griddata(trans, z, (Xg, Yg), method='linear')
    Zg[np.isnan(Zg)]=0
    fig = plt.figure(figsize=(15, 10), dpi= 70)
    ax = fig.gca(projection='3d')
    #surf = ax.plot_surface(x, y, z, cmap=cm.coolwarm,linewidth=0, antialiased=False)
    #ax.plot_surface(x,y,z)
    #ax.plot_wireframe(x, y, z, rstride=10, cstride=10)
    if(way=='s'):
        ax.plot_surface(Xg, Yg, Zg,cmap=cm.coolwarm)
    elif(way=='w'):
        ax.plot_wireframe(Xg, Yg, Zg)
    #ax.scatter(x,y,z,c=z)
    plt.show()

pcaed_data = pca(x_scaled_std)
three_d_viz(pcaed_data,y,'s')
```

```python
class Model(object):
    def __init__(self):
        self.input = tf.placeholder(tf.float32,shape=[None,4],name='input')
        self.label = tf.placeholder(tf.float32,shape=[None,1],name='label')
        self.global_step = tf.Variable(0, name='global_step',trainable=False)
        self.istraining = tf.placeholder(tf.bool,name='istraining')
        
    def dense(self,x, size, scope):
        return tf.contrib.layers.fully_connected(x, size, 
                                                 activation_fn=None,
                                                 scope=scope)

    def dense_batch_relu(self,x, phase, scope):
        with tf.variable_scope(scope):
            h1 = tf.contrib.layers.fully_connected(x, 10, 
                                                   activation_fn=None,
                                                   scope='dense')
            h2 = tf.contrib.layers.batch_norm(h1, 
                                              center=True, scale=True, 
                                              is_training=phase,
                                              scope='bn')
            return tf.nn.relu(h1, 'relu')
        
    def inference(self):
        #dense_out = self.denselayer_with_dropout(self.input,1,1)
        #return dense_out
        h1 = self.dense_batch_relu(self.input, self.istraining,'layer1')
        h2 = self.dense_batch_relu(h1, self.istraining, 'layer2')
        logits = self.dense(h2, 1, 'logits')
        return logits
    
    def build_optimizer(self,loss_val,trainable_var,learning_rate):
        optimizer = tf.train.AdamOptimizer(learning_rate) 
        grads = optimizer.compute_gradients(loss_val, var_list=trainable_var)
        return optimizer.apply_gradients(grads,global_step=self.global_step)
        
    def build_train(self,learning_rate):
        raw_output = self.inference()
        rawloss = tf.losses.mean_squared_error(self.label,raw_output)
        loss = tf.reduce_mean(rawloss)
        trainable_var = tf.trainable_variables()
        train_op = self.build_optimizer(loss,trainable_var,learning_rate)
        return train_op,loss,raw_output
    
    def before_session_initialization(self,learning_rate):
        self.train_op,self.loss_op, self.raw_out = self.build_train(
            learning_rate=learning_rate)
        
    def after_session_initialization(self,sess):
        sess.run(tf.global_variables_initializer())
        sess.run(tf.local_variables_initializer())
        self.train_writer = tf.summary.FileWriter('./', sess.graph)
        
    def train(self,sess,data,label,total_epoch):
        for i in range(total_epoch):
            sys.stdout.write(
                '\r>> training process: {}%'.format(100.0*i/total_epoch)
            )
            sys.stdout.flush()
            batch_data = data
            batch_label = label
            feed_dict = {self.input: batch_data, self.label: batch_label,self.istraining:True}
            _, train_loss,out= sess.run([self.train_op,self.loss_op,self.raw_out], feed_dict=feed_dict)
        print(train_loss)
            #print(out)
            #print(batch_label)
            
    def test(self,sess,data):
        feed_dict = {self.input: data,self.istraining:False}
        out= sess.run([self.raw_out], feed_dict=feed_dict)
        return out
        
    def vb (self,sess):
        print(sess.run(self.w))
        print(sess.run(self.b))

tf.reset_default_graph()
model = Model()
model.before_session_initialization(learning_rate=1e-3)
sess = tf.InteractiveSession()
model.after_session_initialization(sess)
model.train(sess,x_scaled_std,y,5000)
out = model.test(sess,x_scaled_std)
sess.close()

#plt.scatter(x,y)
#print(out[0].shape)
#plt.plot(x,out[0],color='r')
#plt.show()
```

```python
three_d_viz(pcaed_data,np.array(out),'w')
```

```python
def scatter_diff(xy,z1,z2):
    x = xy[:,0].squeeze()
    y = xy[:,1].squeeze()
    z1 = z1.squeeze()
    z2 = z2.squeeze()
    #fig = plt.figure(figsize=(15, 10), dpi= 70)
    #ax = fig.gca(projection='3d')
    #ax.scatter(x,y,z1,color='r',marker='o')
    #ax.scatter(x,y,z2,color='b',marker='^')
    
    xi = np.linspace(min(x), max(x))
    yi = np.linspace(min(y), max(y))
    Xg, Yg = np.meshgrid(xi, yi)
    Zg1 = griddata(xy, z1, (Xg, Yg), method='linear')
    Zg1[np.isnan(Zg1)]=0
    Zg2 = griddata(xy, z2, (Xg, Yg), method='linear')
    Zg2[np.isnan(Zg2)]=0
    
    fig = plt.figure(figsize=(15, 10), dpi= 70)
    ax = fig.gca(projection='3d')

    ax.plot_surface(Xg, Yg, Zg1,cmap=cm.coolwarm)
    ax.plot_wireframe(Xg, Yg, Zg2)
    #ax.scatter(x,y,z,c=z)
    plt.show()

scatter_diff(pcaed_data,y,np.array(out))
```
